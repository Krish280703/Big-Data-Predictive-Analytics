```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

def dist_summary(dist_data, dist_name):
    # Convert distribution data to a pandas Series
    dist_series = pd.Series(dist_data)

    # Plot histogram
    dist_series.hist()
    plt.title(f'Histogram of {dist_name}')
    plt.xlabel('Values')
    plt.ylabel('Frequency')
    plt.show()

    # Return summary statistics
    summary_stats = dist_series.describe()
    return summary_stats
def sim_normal(nums,mean,standard_deviation):
    for n in nums:
        normal_data = np.random.normal(loc = mean, scale=standard_deviation,size = n)
        # Create a distribution name based on the size 'n'
        dist_name = f'Normal(n={n}, mean={mean}, std_dev={standard_deviation})'
        
        # Use the dist_summary function to display summary statistics
        summary_stats = dist_summary(normal_data, dist_name)
        print(f'\nSummary Statistics for {dist_name}:\n{summary_stats}\n{"="*40}')

# Example Usage:
# Assuming you have a distribution named 'example_dist' with data
# Replace 'example_dist_data' with your actual distribution data
example_dist_data = [1, 2, 2, 3, 3, 3, 4, 4, 5]
example_dist_name = 'example_dist'


result_summary = dist_summary(example_dist_data, example_dist_name)
print(result_summary)
```

```python
your_mean = 0
standard_deviation = 1
your_list_of_sizes = [50, 100, 200]

sim_normal(your_list_of_sizes, your_mean, standard_deviation)
```
